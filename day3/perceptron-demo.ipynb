```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(df['cgpa'],df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_   #returns w1, w2
```

```python
p.intercept_   # returns b value
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
#clf is classifier, in this case we used perceptron object p to classify
```
